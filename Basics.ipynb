```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
print(star_wars.shape)
#print(star_wars.head(5))
star_wars = star_wars[star_wars['RespondentID'].isna() == False]
print(star_wars.shape)
```

```python
yes_no_dict = {
            "Yes": True,
            "No": False
}
cols = ['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']
for col in cols:
    star_wars[col] = star_wars[col].map(yes_no_dict)
print(star_wars[col])
```

```python
star_wars.columns
```

```python
import numpy as np
name_bool = { 
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones ": True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}
for i in range(3,9):
    star_wars[star_wars.columns[i]] = star_wars[star_wars.columns[i]].map(name_bool)

name_dict = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
}
star_wars = star_wars.rename(columns=name_dict)
```

```python
star_wars[star_wars.columns[3:9]]
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
name_dict = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1',
    'Unnamed: 10' : 'ranking_2',
    'Unnamed: 11' : 'ranking_3',
    'Unnamed: 12' : 'ranking_4',
    'Unnamed: 13' : 'ranking_5',
    'Unnamed: 14' : 'ranking_6'
}

star_wars = star_wars.rename(columns=name_dict)
star_wars.columns
```

```python
mean_cols = star_wars.mean(axis=0, numeric_only=True)
mean_cols
fig, ax = plt.subplots()
x = np.arange(6)
rank_cols = ['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5', 'ranking_6']
ax.bar(x,mean_cols[rank_cols],width=0.5)
ax.set_xticks(x+0.25)
ax.set_xticklabels(rank_cols, rotation=90)
```

Based on the survey, 'Empire Strikes Back' ranks the best while 'Episode 3: Revenge of the Sith' ranks the worst. If the heading of the survey was actually - 'Does everyone realise that Empire Strikes Back' is really the best of the bunch then maybe the surveyers were biased before they started answering the questions?

```python
seen_cols = ['seen_1','seen_2','seen_3','seen_4','seen_5','seen_6']
x=np.arange(len(seen_cols))
sum_col = star_wars.sum(numeric_only=True)
sum_col.astype(int)
fig, ax = plt.subplots()
ax.bar(x,sum_col[seen_cols])
ax.set_xticks(x+0.3)
ax.set_xticklabels(seen_cols,rotation=90)
```

```python
# ranking movies based on gender
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
males_mean = males.mean(numeric_only=True)
females_mean = females.mean(numeric_only=True)
print("Males - \n",males_mean,"\nFemales - \n",females_mean)
fig, (ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(10,10))
ax1.bar(x,males_mean[rank_cols])
ax2.bar(x,females_mean[rank_cols])
ax1.set_xticks(x)
ax2.set_xticks(x+0.3)
ax2.set_xticklabels(rank_cols)
```
